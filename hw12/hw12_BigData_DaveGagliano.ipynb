```python
import sqlite3
import pandas as pd
```

```python
# connects to a database and create a connection object
connection = sqlite3.connect('ch17/books.db')
```

```python
# return the table names
pd.read_sql_query("""SELECT name 
                  FROM sqlite_master 
                  WHERE TYPE= 'table'
                  """, connection)
```

<h4>a) Select all authors' last names from the authors table in descending order</h4>

```python
# I used two different ways to sort, one with pandas (a way that I knew) and the new way with SQL to compare  

# use SQL query to return a dataframe with the query's results and sort in descending order by last name using pandas
df_authors = pd.read_sql('SELECT * FROM authors', connection)
df_authors_sorted = df_authors.sort_values('last', ascending=False)
df_authors_sorted
```

```python
# use SQL query to return a dataframe with the query's results and sort in descending order by last name using SQL
pd.read_sql("""
            SELECT * 
            FROM authors
            ORDER BY last DESC
            """, connection)
```

<h4>b) Select all book titles from the titles table in ascending order</h4>

```python
# I used two different ways to sort, one with pandas (a way that I knew) and the new way with SQL to compare

# use SQL query to retrun a dataframe with the query's results and sort in ascending order with pandas
df_titles = pd.read_sql('SELECT * FROM titles', connection)
df_titles_sorted = df_titles.sort_values('title')
df_titles_sorted
```

```python
# use SQL query to retrun a dataframe with the query's results and sort in ascending order with SQL
pd.read_sql("""
            SELECT * FROM titles
            ORDER BY title ASC
            """, connection)
```

<h4>c) use an INNER JOIN to select all the books from a specific author. Include the title, copyright year, and ISBN. Order the information alphabetically by title.</h4>

```python
# use SQL query to retrun a dataframe with the query's results
firstname= 'Paul'
lastname = 'Deitel'
pd.read_sql("""
            SELECT * 
            
            FROM authors 

            WHERE first LIKE 'Paul' AND  last LIKE 'Deitel'
            """, connection)
```

```python
# use SQL query to retrun a dataframe with the query's results
firstname= 'Paul'
lastname = 'Deitel'

def select_all_books_author(firstname, lastname):
    
    return pd.read_sql(f"""
                        SELECT 
                        authors.first, 
                        authors.last, 
                        titles.title, 
                        titles.copyright, 
                        titles.isbn
           
                        FROM authors
                
                        INNER JOIN author_ISBN
                            ON authors.id = author_ISBN.id
            
                        INNER JOIN titles
                            ON author_ISBN.isbn = titles.isbn
                
                        WHERE authors.first LIKE '{firstname}' AND  authors.last LIKE '{lastname}'
                
                        ORDER BY title ASC
                        """, connection)

#display output in Jupyter Notebook
select_all_books_author(firstname, lastname)
```

```python
# display output in program
print(select_all_books_author(firstname, lastname))         
```
